```python
from datasets import load_metric
metric = load_metric("rouge")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['model', 'precision', 'recall', 'rougeL f1'])

for i in range(26):
    model_num = str(i+1).zfill(2)
    pred_file = 'predictions_model'+model_num+'.csv'
    model_predictions = pd.read_csv(pred_file)['Generated Text']
    gold_references = pd.read_csv(pred_file)['Actual Text']
    final_score = metric.compute(predictions=model_predictions, references=gold_references)
    precision = final_score['rougeL'].mid.precision
    recall = final_score['rougeL'].mid.recall
    try:
        rougeL_f1 = 2 * precision * recall / (precision + recall)
    except:
        rougeL_f1 = 'und'
    print('model'+model_num+':')
    df.loc[len(df.index)] = ['model'+model_num, precision, recall, rougeL_f1]
#     print(final_score)
    print('\t','rougeL_f1: ', rougeL_f1)
    print('\t','precision: ', precision, '\t','recall: ', recall)
    with open('./scores/scores_model'+str(model_num)+'.txt', 'w') as writefile:
        writefile.write(str(final_score))
```

```python
df.to_csv('./scores/scores.csv', index=False)
```
